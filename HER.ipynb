```python
import matplotlib.pyplot as plt
from bbrl_gym.envs.maze_mdp import MazeMDPEnv
import gym
import bbrl_gym
from algo.QL_HER import QLearingHER
```

```python
env = gym.make("MazeMDP-v0", kwargs={"width": 5, "height": 5, "ratio": 0.2})
env.reset()
env.init_draw("The maze")
env.mdp.plotter.using_notebook = False
algo = QLearingHER(
    env,
    max_episodes=2500,
    her_k=5,
    batch_size=400,
    buffer_size=1000,
    max_steps=50,
    epsilon=0.05,
)
q_list, time_list = algo.train()

plt.plot(range(len(q_list)), q_list)
plt.xlabel("Episodes")
plt.ylabel("Q-Norm")
plt.show()
plt.figure(figsize=(18, 5))
plt.plot(range(len(time_list)), time_list)
plt.xlabel("Episodes")
plt.ylabel("Steps to goal")
plt.show()
algo.render_policy()
```
